# Importando libs e lendo o dataset

```python
import pandas as pd
import kagglehub
```

```python
def return_file(file_path):
   from pathlib import Path

   for file in Path(file_path).iterdir():
      if file.is_file():
         return file
```

```python
# Download latest version
path = kagglehub.dataset_download("lfarhat/brasil-students-scholarship-prouni-20052019")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(f"{str(path/return_file(path))}")
df
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
lista_colunas = df.columns
print(lista_colunas)
```

```python
lista_renomeada = [col.upper().replace('_BOLSA', '').replace('NOME_', '').replace('_CURSO_', '').replace('BENEFICIARIO_', '').replace('_BENEFICIARIO', '').replace('_CURSO', '').replace('CURSO_', '').replace('SIGLA_', '').replace('IES', 'INSTITUICAO') for col in lista_colunas]
print(lista_renomeada)
```

```python
df_renamed = df.copy()
df_renamed.rename(columns=dict(zip(lista_colunas, lista_renomeada)), inplace=True)
df_renamed.columns
```

```python
df_renamed.isna().sum()
```
